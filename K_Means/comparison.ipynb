 ## Comparison Notebook
 This notebook is intended to compare implementation from scratch with common libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

Using scikit-learn:

```python
data = pd.read_csv("data.csv")
X = data[['x', 'y']].values
kmeans = KMeans(n_clusters=3, init = 'random', n_init = 10)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.title('K-means Clustering Using scikit-learn')
plt.grid(True)
plt.show()
```

Implementation from Scratch:

```python
from_scratch = pd.read_csv("output.csv")
X_scratch = from_scratch[['cluster', 'x', 'y']].values
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scratch[:, 1], X_scratch[:, 2], c=X_scratch[:, 0], cmap='viridis', s=50, alpha=0.5)
plt.title('K-means Clustering (Implementation from Scratch)')
plt.grid(True)
plt.show()
```
